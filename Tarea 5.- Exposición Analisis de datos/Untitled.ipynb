#Análisis Exploratorio de Datos.
Se refiere a el arte de observar cuidadosamente la forma estructurada de los datos.

Existen 4 conceptos claves en la exploración de los datos.
1. Relevancia.- Visualización de los datos.
2. Residuales.- Diferencia entre los valores observados de una variable y su predicción con algún modelo matemático.
3. Re-Expresión.- La explicación de transformaciones matemñaticas para una o mas variables.
4. Resistencia.- La habilidad de una caracterización de datos para evitar la influencia desproporcionada de datos anormales.

##Anomalias en los Datos.
###Valores Atípicos y su Influencia.
El significado que se le da a un valor atípico es el tomado por Barnett y Lewis.
> Una observación (o conjunto de observaciones) que aparentan ser inconsistentes con el resto del conjunto de datos.

###Detectar valores atípicos univariantes
Se refiere al proceso de identificar los valores atipicos en una sola variable. En este caso, "la inconsitencia con el resto de los datos" se interpreta como "inusualmente muy por encima o por debajo de los valores tipicos de los datos".

Se hablara de 3 diferentes procesos para la detección de valores atipicos invariables. Usando los siguientes elemento:
- Valor Tipico.
- Disperción de los datos.
- Umbral.

Los procesos son los siguientes.
1. La regla de edicion de 3 sigma
    - Como valor tipico; la media.
    - Como distribuidor de datos; la desviación estandar.
    - Como umbral ; 3 desviaciones estandar.

2. El identificador de Hampel
    - Como valor tipico; La mediana.
    - Como distribuidor de de datos; El estimador de scala MADM.
    - Como umbral; 3 Estimadores de scala MADM.
    
    EL MADM representa una alternativa resistente al valor atipico.
    
3. La regla de valores atipicos de la gráfica de caja.
    - EL valor tipico mas alto; upper quartile Xu.
    - EL valor tipico mas pequeño; lower quartile Xl.
    - EL distribuidor de datos; interquartile distance q = Xu - Xl.
